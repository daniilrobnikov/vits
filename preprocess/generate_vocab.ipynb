```python
def get_non_empty_content(md_table):
    non_empty_content = []

    # Split table by lines
    lines = md_table.split("\n")

    for line in lines:
        # Split each line by "|" to get the cells
        cells = line.split("|")

        for cell in cells:
            cell_content = cell.strip()

            # If the cell content is not empty, add it to the list
            if cell_content != "":
                non_empty_content.append(cell_content)

    non_empty_content = "".join(non_empty_content)

    # unique non_empty_content
    non_empty_content = set(non_empty_content)
    non_empty_content = "".join(non_empty_content)

    # sort non_empty_content
    non_empty_content = sorted(non_empty_content)

    return non_empty_content


# Consonants
consonants = get_non_empty_content(consonants)
consonants_other = get_non_empty_content(consonants_other)
manner_of_articulation = get_non_empty_content(manner_of_articulation)
# Vowels
vowels = get_non_empty_content(vowels)
vowels_other = get_non_empty_content(vowels_other)
# Diacritics
articulation = get_non_empty_content(articulation)
air_flow = get_non_empty_content(air_flow)
phonation = get_non_empty_content(phonation)
rounding_and_labialization = get_non_empty_content(rounding_and_labialization)
syllabicity = get_non_empty_content(syllabicity)
consonant_release = get_non_empty_content(consonant_release)
co_articulation = get_non_empty_content(co_articulation)
tongue_root = get_non_empty_content(tongue_root)
fortis_and_lenis = get_non_empty_content(fortis_and_lenis)
# Suprasegmentals
stress = get_non_empty_content(stress)
length = get_non_empty_content(length)
rhythm = get_non_empty_content(rhythm)
tones = get_non_empty_content(tones)
intonation = get_non_empty_content(intonation)
```

```python
# All symbols
_ipa = (
    consonants
    + consonants_other
    + manner_of_articulation
    + vowels
    + vowels_other
    + articulation
    + air_flow
    + phonation
    + rounding_and_labialization
    + syllabicity
    + consonant_release
    + co_articulation
    + tongue_root
    + fortis_and_lenis
    + stress
    + length
    + rhythm
    + tones
    + intonation
)

print(_ipa)
print(len(_ipa))
```
